```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('company_sales_data.csv')

# Exercise 1: Total profit of all months shown using a line plot
plt.figure(figsize=(10, 5))
plt.plot(data['month_number'], data['total_profit'], marker='o', linestyle='-', color='b', label='Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit per Month')
plt.xticks(data['month_number'])
plt.legend()
plt.show()

# Exercise 2: Total profit of all months with style properties
plt.figure(figsize=(10, 5))
plt.plot(data['month_number'], data['total_profit'], linestyle='--', color='red', marker='o', linewidth=3, label='Profit Data')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend(loc='lower right')
plt.title('Total Profit of All Months')
plt.show()

# Exercise 3: All product sales data using a multiline plot
plt.figure(figsize=(10, 5))
for product in data.columns[1:-2]:  
    plt.plot(data['month_number'], data[product], marker='o', label=product)
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.title('Sales Data')
plt.legend()
plt.show()

# Exercise 4: Toothpaste sales data of each month shown using a scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(data['month_number'], data['toothpaste'], label='Toothpaste Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Toothpaste Sales Data')
plt.grid(True, linestyle='--')
plt.legend()
plt.show()

# Exercise 5: Face cream and facewash product sales data using the bar chart
plt.figure(figsize=(10, 5))
bar_width = 0.35
index = data['month_number']
plt.bar(index, data['facecream'], bar_width, label='Face Cream Sales Data', align='center')
plt.bar(index + bar_width, data['facewash'], bar_width, label='Face Wash Sales Data', align='center')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.title('Sales Data of Facewash and Facecream')
plt.xticks(index + bar_width / 2, data['month_number'])
plt.legend()
plt.show()

# Exercise 6: Bathing soap of all months and save the plot to disk
plt.figure(figsize=(10, 5))
plt.bar(data['month_number'], data['bathingsoap'], label='Bathing Soap Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.title('Bathing Soap Sales Data')
plt.legend()
plt.savefig('bathing_soap_sales_data.png')
plt.show()

# Exercise 7: Total profit of each month shown using the histogram
plt.figure(figsize=(10, 5))
plt.hist(data['total_profit'], bins=[150000, 200000, 250000, 300000, 350000], rwidth=0.8, color='green', alpha=0.7)
plt.xlabel('Profit Range')
plt.ylabel('Frequency')
plt.title('Profit Data')
plt.show()

# Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
plt.figure(figsize=(8, 8))
total_sales_products = data.iloc[:, 1:-2].sum()
plt.pie(total_sales_products, labels=total_sales_products.index, autopct='%1.1f%%', startangle=140)
plt.title('Sales Data')
plt.show()

# Exercise 9: Bathing soap and facewash of all months using the Subplot
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
ax[0].plot(data['month_number'], data['bathingsoap'], label='Bathing Soap Sales', color='skyblue', marker='o')
ax[1].plot(data['month_number'], data['facewash'], label='Face Wash Sales', color='orange', marker='o')
ax[0].set_title('Bathing Soap Sales Data')
ax[1].set_title('Face Wash Sales Data')
ax[0].set_xlabel('Month Number')
ax[1].set_xlabel('Month Number')
ax[0].set_ylabel('Sales Units')
ax[1].set_ylabel('Sales Units')
ax[0].legend()
ax[1].legend()
plt.tight_layout()
plt.show()

# Exercise 10: All product sales data using a stack plot
plt.figure(figsize=(10, 5))
product_data = data.iloc[:, 1:-2].T.apply(lambda x: x.tolist(), axis=1).tolist()
plt.stackplot(data['month_number'], product_data, labels=data.columns[1:-2])
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.title('All Product Sales Data Using Stack Plot')
plt.legend(loc='upper left')
plt.show()
```
